## Wyrażenia listowne

```python
liczby = [1, 2, 3, 4, 5, 6]

potegi_dwojki = []

for liczba in liczby:
    potega = 2 ** liczba
    potegi_dwojki.append(potega)
```

```python
[liczba**2 
 for liczba in liczby]
```

```python
[liczba 
 for liczba in liczby
 if liczba % 2 == 0]
```

```python
[    
    i ** 2
    for i in range(11)
]
```

## Wyrażenie generujące (generators)

```python
parzyste1 = [liczba 
 for liczba in range(5000)
 if liczba % 2 == 0]
```

```python
parzyste2 = (liczba 
 for liczba in range(5000)
 if liczba % 2 == 0)
# nie mamy wszystkich od razu, generujemy na bieżąco, zatem zajmujemy tyko troszeczkę miejsca w pamięci!
```

```python
import sys
print(sys.getsizeof(parzyste1))
print(sys.getsizeof(parzyste2))
```

```python
sum(parzyste2)
```

## Wyrażenie słownikowe

```python
imiona = ['Jan', 'Franciszek', 'Kacper']
{
    name : len(name)
    for name in imiona
}
```

## Wyrażenia zbioru

```python
names = {'arkadiusz', 'Wioletta', 'Marek', 'monika'}

names2 = {
            name.capitalize() 
            for name in names
         }

print(names2)
```

### Zadanie
Znajdź liczby od 100 do 470 włącznie, które są podzielne przez 7 ale niepodzielne przez 5

```python
solution = (
    liczba 
    for liczba in range(100, 471)
    if liczba % 7 == 0
    if liczba % 5 != 0
)
for number in solution:
    print(number)
```
