### Liczenie powierzchni figur

```python
def calculate_area():
    figure = input('Wybierz: prostokąt, kwadrat, trójkąt, trapez, koło').strip().lower()

    if figure == 'prostokąt':
        a = float(input('Podaj pierwszy bok'))
        b = float(input('Podaj drugi bok'))
        area = a*b
        return area
    
    if figure == 'kwadrat':
        a = float(input('Podaj bok'))
        area = a**2
        return area
    
    if figure == 'trójkąt':
        a = float(input('Podaj bok'))
        h = float(input('Podaj wysokość'))
        area = (a*h)/2
        return area
    
    if figure == 'trapez':
        a = float(input('Podaj pierwszy bok'))
        b = float(input('Podaj drugi bok'))
        area = a*b
        return area
    
    if figure == 'koło':
        a = float(input('Podaj promień'))
        area = 2*3.14 * a
        return area
```

### Sumowanie dodatnich liczb w liście

```python
def suma_dodatnich_elementow(x):
    mk = (int(i) for i in x if i > 0)
    return sum(mk)
```

```python
lista = range(-91, 128, 7)
print(lista)
suma_dodatnich_elementow(lista)
```
