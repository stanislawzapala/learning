Time module

```python
import time

def sumuj_do_n(n):
    suma = 0
    for i in range(1, n + 1):
        suma += i
    return suma

def sumuj_do_n_szybko(n):
    return n * (n + 1) // 2

def sumuj_do_n2(n):
    if n != 0:
        suma = n + sumuj_do_n2(n - 1)
    else:
        suma = 0
    return suma


def finish_time(start_time):
    end_time = time.perf_counter()
    elapsed_time = end_time - start_time
    print(f"Czas wykonania: {elapsed_time:.6f} sekund")

n = 1000
print(f"Sumowanie do {n} za pomocą pętli:")
start_time = time.perf_counter()
wynik = sumuj_do_n(n)
finish_time(start_time)
print(f"Wynik: {wynik}")

print(f"\nSumowanie do {n} za pomocą wzoru:")
start_time = time.perf_counter()
wynik = sumuj_do_n_szybko(n)
finish_time(start_time)
print(f"Wynik: {wynik}")

print(f"\nSumowanie do {n} za pomocą rekurencji:")
start_time = time.perf_counter()
wynik = sumuj_do_n2(n)
finish_time(start_time)
print(f"Wynik: {wynik}")
```
