```python
# ==========================================
# BATTLECODE 2026 - TERMINALOWY DEKODER MARKERÓW
# ==========================================

# Słowniki (Odwzorowanie z Twojego kodu)
INT_TO_ENV = {
    0: "EMPTY", 
    1: "WALL", 
    2: "ORE_TITANIUM", 
    3: "ORE_AXIONITE"
}

INT_TO_ENTITY = {
    1: "BUILDER_BOT", 2: "CORE", 3: "GUNNER", 4: "SENTINEL", 
    5: "BREACH", 6: "LAUNCHER", 7: "CONVEYOR", 8: "SPLITTER", 
    9: "ARMOURED_CONVEYOR", 10: "BRIDGE", 11: "HARVESTER", 
    12: "FOUNDRY", 13: "ROAD", 14: "BARRIER", 15: "MARKER"
}

def decode_marker(value: int):
    """Rozbija 32-bitowego inta na czytelne informacje."""
    
    # 1. Dekodowanie bitowe
    team_val = value & 0b1
    entity_val = (value >> 1) & 0b1111
    env_val = (value >> 5) & 0b11
    y = (value >> 7) & 0b111111
    x = (value >> 13) & 0b111111
    turn = (value >> 19) & 0b11111111111
    msg_type = (value >> 30) & 0b11
    
    # 2. Tłumaczenie na nazwy
    is_enemy = "WRÓG" if team_val == 1 else "SOJUSZNIK"
    b_type = INT_TO_ENTITY.get(entity_val, "BRAK BUDYNKU")
    env = INT_TO_ENV.get(env_val, "NIEZNANE")
    
    # 3. Wyświetlanie wyników
    print("\n" + "="*40)
    print(f"DEKODOWANIE MARKERA: {value}")
    print("="*40)
    
    if msg_type == 0:
        print("TYP WIADOMOŚCI : [0] Aktualizacja Mapy")
        print(f"TURA ZNALEZIENIA: {turn}")
        print(f"POZYCJA (X, Y)  : ({x}, {y})")
        print(f"TEREN           : {env}")
        
        if entity_val == 0:
            print(f"BUDYNEK         : {b_type}")
        else:
            print(f"BUDYNEK         : {b_type} ({is_enemy})")
            
    elif msg_type == 1:
        # Rozpakowywanie Kanału 1 (Rozkaz Taktyczny)
        payload = value & 0b1111111
        priority = payload & 0b111
        command_id = (payload >> 3) & 0b1111
        
        print("TYP WIADOMOŚCI : [1] Rozkaz Taktyczny")
        print(f"TURA WYDANIA    : {turn}")
        print(f"CEL (X, Y)      : ({x}, {y})")
        print(f"ID KOMENDY      : {command_id}")
        print(f"PRIORYTET       : {priority}")
    else:
        print(f"NIEZNANY TYP WIADOMOŚCI: {msg_type}")
        
    print("="*40 + "\n")

# Pętla nasłuchująca w terminalu
if __name__ == "__main__":
    print("Uruchomiono dekoder markerów. Wpisz 'q' aby wyjść.")
    while True:
        user_input = input("Podaj wartość markera (int): ")
        if user_input.lower() in ['q', 'quit', 'exit']:
            break
        
        try:
            val = int(user_input)
            decode_marker(val)
        except ValueError:
            print("Błąd: Podaj poprawną liczbę całkowitą (integer)!")
```
