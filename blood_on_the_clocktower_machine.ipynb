```python
# List of roles
imp_roles = [
    # Demon
    'Pijawka',

    # Minion
    'Cerenovus',
    'Asasyn',
    'Adwokat diabła',

    # Outsiders
    'Mutant',

    # Important Townsfolk
    'Uczony',
    'Amnezjak',
    'Kanibal'
]

other_roles = [
    'Szlachcic',
    'Babcia',
    'Mnich',
    'Farmer',
    'Chłopek roztropek',
    'Chłopek roztropek'

]
```

```python
# List of players
exp_players =[
    'Ignacy',
    'Olek',
    'Gucio',
    'Aga Rosińska',
    'Artur',
    'Gabi',
    'Grzesiek',
    'Kuba H',
    'Michał',
    'Karolina',
    'Julianka'
]

new_players =[
    'Asia Sko',
    'Agata Płatek',
    'Yana'
]
```

```python
import random
import pandas as pd

def randomize_non_randomly(e_p = exp_players, n_p = new_players, i_r = imp_roles, o_r = other_roles):

    shuffled_players = random.sample(e_p, len(e_p)) + random.sample(n_p, len(n_p))
    print(f'Number of players: {len(shuffled_players)}.')

    ordered_roles = i_r + o_r
    shuffled_roles = random.sample(i_r, len(i_r)) + random.sample(o_r, len(o_r))
    print(f'Number of roles: {len(shuffled_roles)}.')

    final_list = list(zip(shuffled_players, shuffled_roles))
    final_list = sorted(final_list, key=lambda x: ordered_roles.index(x[1]))
    
    df = pd.DataFrame(final_list, columns=['Player', 'Role'], index=range(1, len(final_list) + 1))
    return df
```
