# G.71 

```python
def f(x):
    if x==0: return 1
    cnt = 0
    while x != 0:
        x //= 10
        cnt+=1
    return cnt
```

```python
f(10) #Złożoność O(n), bo przechodzimy przez wszystkie n w x
```

# G. 72

```python
def f(x):
    positions_dict = {}

    for i in range(len(x)):
        if x[i] not in positions_dict:
            positions_dict[x[i]] = i
        else:
            first_idx = positions_dict[x[i]]
            positions_dict[x[i]] = i - first_idx
    return positions_dict
```

```python
def f(x):
    positions_dict = {}

    for i, val in enumerate(x):
        if x[i] not in positions_dict:
            positions_dict[val] = i
        else:
            positions_dict[val] = i - positions_dict[val]
    return positions_dict
```

# G.74

```python
def mergesort(t):
    if len(t) <= 1:
        return t
        
    middle_idx = len(t) // 2
    left = mergesort(t[:middle_idx])
    right = mergesort(t[middle_idx:])
    
    wynik = []
    # Zamiast pop(), używamy indeksów i oraz j
    i = 0
    j = 0
    
    # Scalamy dopóki nie wyczerpiemy jednej z list
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            wynik.append(left[i])
            i += 1
        else:
            wynik.append(right[j])
            j += 1
            
    # Gdy jedna lista się skończy, doklejamy resztę z drugiej.
    # Wycinki (slices) w Pythonie są tutaj w porządku, 
    # bo to jednorazowa operacja skopiowania reszty.
    wynik.extend(left[i:])
    wynik.extend(right[j:])
    
    return wynik
```

```python
mergesort([15,10,5,9,6,1,2,8,3,4,11])
```
