# Lab G.41:
Napisz funkcję, która przyjmuje jako parametry długości 3 odcinków i sprawdzi, czy z odcinków takiej długości można zbudować trójkąt

```python
def check_if_triangle(a,b,c):
    mk = (a,b,c)
    # Dla każdego boku sprawdzamy, czy jest on większy lub równy pozostałym dwóm bokom
    # Jeśli tak - to nie da się zbudować trójkąta.
    for i in range(3):
        if mk[i] >= mk[(i + 1) % 3] + mk[(i + 2) % 3]:
            print('Niestety nie da się zbudować trójkąta z tych boków')
            return
    print('Da się zbudować trójkąt z tych boków')
    return
```

```python
check_if_triangle(1,2,3)
check_if_triangle(2,2,3)
```

# Lab G.42:
Napisz funkcję, która jako parametr przyjmuje listę list i łączy te listy, tzn zwraca listę zawierającą kolejne elementy wszystkich tych list.

```python
def add_lists(x):
    # Inicjujemy naszą listę, do której będziemy dołączać kolejne listy
    big_list = []
    # Po kolei dołączamy każdą listę
    for i in x:
        big_list += i
    return big_list
```

```python
add_lists([[1,2,3,4], [4,1,2], [8,432,92,94,4]])
```

# Lab G.43:
Napisz funkcję, która przyjmuje jako argumenty dwie listy: listę liczb i dzielników. Zakładamy, że na obu listach znajdują się liczby naturalne i druga z nich nie zawiera zera. Zadaniem funkcji jest zwrócenie słownika, który dzielnikom przypisuje liczby przez nie podzielne z pierwszej listy. Kluczami słownika powinny być liczby z listy dzielników, natomiast odpowiadającymi im wartościami listy liczb przez nie podzielnych, znajdujących się na pierwszej liście. Jeśli dla danego dzielnika nie ma żadnej liczby przez niego podzielnej na pierwszej liście, ma mu zostać przypisana pusta lista.

```python
def division_lister(num, div):
    div_dict = {}
    
    # Dla każdego dzielnika domyślnie przypisujemy pustą listę
    for d in div:
        div_dict[d] = []

        for n in num:
            # Jeśli liczba jest podzielna przez dany dzielnik, dodajemy ją do listy naszego dzielnika
            if n % d == 0:
                div_dict[d].append(n) 
    return div_dict
```

```python
# Test
numbers = [2, 4, 6, 8, 10, 15, 20, 432, 343, 566]
divisors = [2, 3, 5, 99]

division_lister(numbers, divisors)
```

# Lab G.44:
Przypuśćmy, że w słowniku został zapisany wielomian (wzór), w taki sposób, że zapisano w nim pary(i, c_i) dla niezerowych współczynników c_i. Na przykład, wielomianowi x^3 + 2 odpowiadałby słownik {3:1.0, 0:2.0}. Napisz funkcję, która przyjmuje jako parametr zapisany w ten sposób wielomian oraz wartość zmiennej x i zwraca wartość wielomianu w(x).

```python
def calculate_polynomial(w, x):
    # Inicjujemy wartość wielomianu - na początku 0
    mk = 0
    for degree, coef in w.items():
        # Dodajemy wartość przy danej potędze wielomianu 
        mk += coef * x ** degree
    return mk
```

```python
# Wielomian: 5x⁵ + 3x³ + 2x² - x + 7
w = {5: 5, 3: 3, 2: 2, 1: -1, 0: 7}
x = 3

calculate_polynomial(w, x)
```

# Lab G.45:
Napisz funkcję, która przyjmuje 2 słowniki w formacie określonym w zadaniu poprzednim i zwraca słownik odpowiadający sumie tych wielomianów

```python
def sum_polynomials(w1, w2):
    # Sprawdzamy najwyższy stopień tych wielomianów
    max_degree = max(max(w1.keys()), max(w2.keys()))
    result = {}

    for i in range(max_degree + 1):
        # Pobieramy współczynnik przy danej potędze z obu wielomianów, 
        # jeśli nie ma - to 0
        if i in w1:
            coef1 = w1[i]
        else:
            coef1 = 0
        if i in w2:
            coef2 = w2[i]
        else:            
            coef2 = 0

        # Sumujemy te współczynniki i zapisujemy do wyniku
        result[i] = coef1 + coef2
    return result
```

```python
w1 = {5: 5, 4: 2, 3: 3, 2: 2, 1: -1, 0: 7}
w2 = {6: 2, 4: -4, 2: 1, 0: -3}
sum_polynomials(w1, w2)
```

# Lab G.46:
Napisz funkcję, która dla podajen listy licz znajduje i zwraca najdłuższy zawarty w niej niemalejący ciąg liczb, zajmujący kolejne miejsca w tej tablicy. Jeśli takich ciągów jest więcej niż 1, procedura może zwrócić dowolny z nich.

```python
def find_longest_non_decreasing_subsequence(x):
    # Inicjujemy listy do trzymania dotychczas znalezionego ciągu oraz obecnego ciągu
    longest = []
    current = []

    last_num = 0
    for i in x:
        # Jeśli element jest większy lub równy od poprzedniego, to dodajemy go do obecnego ciągu
        if i >= last_num:
            current.append(i)
        else:
            current = [i]
        # Jeśli obecny ciąg jest dłuższy lub równy od dotychczas znalezionego, to aktualizujemy najdłuższy ciąg
        if len(current) >= len(longest):
            longest = current
        # Aktualizujemy ostatnią liczbę
        last_num = i
    
    return longest
```

```python
x = [1, 2, 3, 2, 5, 6, 7, 8, 9]
find_longest_non_decreasing_subsequence(x)
```
