# Wyszukiwanie na liście

```python
def szukaj2(x, lista):
    for i, el in enumerate(lista):
        if (el == x):
            return i
    return None
```

# Ale na posortowanej liście... 
można dużo lepiej! - jak Bolzano-Weierstrass

```python
def bisekcja(x, lista):
    a = 0
    b = len(lista)

    while(a < b):
        c = (a+b) // 2
        
        if (lista[c] == x):
            return c
        elif (lista[c] > x):
            b = c
        else:
            a = c+1

    return None
```

```python
l = np.random.randint(-20,30, size=100).tolist()
l.sort()
print(l)
```

```python
print(bisekcja(-18,l))
```

Def. Niech g:N->R, powiemy, że f:N->R jest O(g(n)) gdy n->\infty jeśli istnieją stałe c>0 oraz n \in N takie, że dla każdego n>N zachodzi f(N)< c*g(n)

W uproszczeniu: chcemy po prostu przybliżyć funkcję najszybciej rosnącym czynnikiem (czyli logn, n, n^2)

## Jeszcze dwie definicje:

Def. Niech g:A->R (A \subset R^k). Powiemy, że f:A->R jest O(g(x)) gdy x to infty, jeśli istnieje c>0 i n>0 t, że forall x in A takiego, że istniejes_i >= N |f(x)| < c |g(x)|
