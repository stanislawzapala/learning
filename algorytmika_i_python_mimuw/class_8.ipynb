G.81

```python
def proc(a,n):
    y = []
    for x in a:
        assert type(x) is int
        z = 1
        for i in range(n):
            z *= x
        y.append(z)
    return y
```

G. 82

```python
def proc(tab):
    a = []
    b = []
    for e in tab:
        assert type(e) is int
        if(e<0):
            a.append(e)
        else:
            b.append(e)
    return (sum(a), sum(b))
```

```python
proc([1,2,3,-1,-2,-3])
# Złożoność pamięciowa: O(n) - musimy przechować dwie listy o łącznej długości n
# Złożoność czasowa: O(n) - musimy przejść przez wszystkie elementy tablicy, aby je rozdzielić i zsumować
```

# G.83

```python
def program(a,b):
    x = []
    i = 0
    j = 0
    min_a = a[i]
    min_b = b[j]
    
    while i < len(a) and j < len(b):
        
        if min_a == min_b:
            x.append(min_a)
            i += 1
            j += 1
            if i < len(a):
                min_a = a[i]
            if j < len(b):
                min_b = b[j]

        elif min_a < min_b and len(a) > 0:
            x.append(min_a)
            i += 1
            if i < len(a):
                min_a = a[i]
        
        elif min_a > min_b and len(b) > 0:
            x.append(min_b)
            j += 1
            if j < len(b):
                min_b = b[j]
    
    if i < len(a):
        x.extend(a[i:])
    if j < len(b):
        x.extend(b[j:])

    return len(x)
```

```python
program([1,2,3,4,5], [2,3,4,5,6])
```
