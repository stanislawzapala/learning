# Zadanie H.I.26
Litery słowa w języku polskim zastępujemy kolejnymi liczbami zgodnie z następującym słownikiem:
{1:'a', 2:'ą',...},
ponadto niech znak spacji będzie kodowany za pomocą liczby 100.
Napisz funkcję encode i decode, które odpowiednio zamieniają napisy na tablice liczb oraz tablice liczb na napisy. Przetestuj na zdaniu "Ala ma kota".

```python
slownik = {1: 'a', 2: 'ą', 3: 'b', 4: 'c', 5: 'ć', \
    6: 'd', 7: 'e', 8: 'ę', 9: 'f', 10: 'g', 11: 'h', \
    12: 'i', 13: 'j', 14: 'k', 15: 'l', 16: 'ł', \
    17: 'm', 18: 'n', 19: 'ń', 20: 'o', 21: 'ó', \
    22: 'p', 23: 'r', 24: 's', 25: 'ś', 26: 't', \
    27: 'u', 28: 'w', 29: 'y', 30: 'z', 31: 'ź', 32: 'ż'}
slownik[100] = ' ' # dodanie wartosci dla spacji
```

```python
# Sposob z odwroceniem slownika na poczatku - najwygodniej
slownik_inverted = {}
for key, value in slownik.items():
    slownik_inverted[value] = key



def encode(x, slownik_do_zamian):
    encoded_x = []
    for i in x:
        try:
            encrypted_i = slownik_do_zamian[i.lower()]
            encoded_x.append(encrypted_i)
        except KeyError:
            return None
    return encoded_x

def decode(x, slownik_do_zamian):
    decoded_x = []
    for i in x:
        decrypted_i = slownik_do_zamian[i]
        decoded_x.append(decrypted_i)

    # potrzebujemy jeszcze metody na polaczenie listy w stringa
    # Sposob 1 - najszybszy:
    # decoded_x_str = ''.join(decoded_x)
    
    # Sposob 2 'na piechote':
    decoded_x_str = ''
    for i in decoded_x:
        decoded_x_str += i

    return decoded_x_str



x = 'Ala ma kota'
x_encoded = encode(x, slownik_inverted)
x_decoded = decode(x_encoded, slownik)

print(x)
print(x_encoded)
print(x_decoded)
```

# Zadanie H.I.27
Litery zastępujemy słownikiem (jak wyżej), a następnie liczymy iloczyn ich kluczy. Znajdź słowo słownikowe, którego iloczyn kluczy będzie najbliżej miliona.

```python
slownik = {1: 'a', 2: 'ą', 3: 'b', 4: 'c', 5: 'ć', \
    6: 'd', 7: 'e', 8: 'ę', 9: 'f', 10: 'g', 11: 'h', \
    12: 'i', 13: 'j', 14: 'k', 15: 'l', 16: 'ł', \
    17: 'm', 18: 'n', 19: 'ń', 20: 'o', 21: 'ó', \
    22: 'p', 23: 'r', 24: 's', 25: 'ś', 26: 't', \
    27: 'u', 28: 'w', 29: 'y', 30: 'z', 31: 'ź', 32: 'ż'}
slownik[100] = ' ' # dodanie wartosci dla spacji
```

```python
slownik_inverted = {}
for key, value in slownik.items():
    slownik_inverted[value] = key

def encode(x, slownik_do_zamian):
    encoded_x = []
    for i in x:
        try:
            encrypted_i = slownik_do_zamian[i.lower()]
            encoded_x.append(encrypted_i)
        except KeyError:
            return None
    return encoded_x

def multiplicate_list(x):
    val = 1
    for i in x:
        val = val * i
    return val

x = 'swędź'
x_encoded = encode(x, slownik_inverted)
multiplication_val = multiplicate_list(x_encoded)

print(x)
print(x_encoded)
print(multiplication_val)
```

# Poszukiwanie słowa za milion
Wykorzystując Słowosieć (Wordnet)

```python
filepath = 'slowosiec.txt'
with open(filepath, 'r', encoding='utf-8') as file:
    slowa_za_milion = []
    current_distance = 1000000
    for line in file:
        word = line.strip()
        if not word: # pomijamy ewentualne puste linijki
            continue

        word_encoded = encode(word, slownik_inverted)
        if word_encoded is None:
            continue
        val = multiplicate_list(word_encoded)
        new_dist = abs(val - 1000000)
        if  new_dist <= current_distance:
            print(f'Mamy nowe slowo za milion! Jest to: {word}')
            slowa_za_milion.append(word)
            current_distance = new_dist
    print(f'Słowo, którego iloczyn jest najbliżej miliona to {slowa_za_milion[-1]} z odległością {current_distance}')
```
