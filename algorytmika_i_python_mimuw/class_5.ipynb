# Lab G.52
Sito Eratostenesa

```python
def sito_erat(x):
    if not x:
        return []

    pierwsza = x[0]
    new_x = []

    for j in x[1:]:
        if j % pierwsza != 0: 
            new_x.append(j)

    return [pierwsza] + sito_erat(new_x)

liczby = list(range(2, 201))
print(sito_erat(liczby))
```

# Lab G.53

```python
def fun(x):
    lista_licznikow = []
    if x == []:
        return 0
    elif type(x) != list:
        return 0
    else:
        for i in x:
            lista_licznikow.append(fun(i))

    return max(lista_licznikow) +1
```

```python
fun([1, [1, 2, []]])
```

```python
fun([[[1, 2], [[1, 3], [1,1]], [2,1]], [1,2]])
```

# Lab G.54

```python
def odwroc(x):
    for i in range(len(x)):
        nowe_i = x.pop(-1)
        x.insert(i, nowe_i)
    return x
```

```python
x = [1,2,3,4,5,6,7,8,9]
odwroc(x)
```

```python
def drukuj_permutacje(x):
    if type(x) == int:
        lista = []
        for i in range(x+1):
            lista.append(i)
    elif len(x) == 1:
        return x
    else:
        lista = x

    permutacje = []

    for j in lista:
        pierwszy_element = j
        mniejsza_lista = lista.copy()
        mniejsza_lista.remove(j)
        lista_mniejszych_permutacji = drukuj_permutacje(mniejsza_lista)
        print(f'Lista mniejszych permutacji dla {j} to {lista_mniejszych_permutacji}')
        for k in lista_mniejszych_permutacji:
            if type(k) != list:
                permutacje.append([j] + [k])
            if type(k) == list:
                permutacje.append([j] + k)

    return permutacje
```

```python
drukuj_permutacje(2)
```
