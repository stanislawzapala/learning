Definiując funkcje w Pythonie określa się
tylko nazwy ich argumentów. Używane typy zależą od tego,
jakie wartości zostaną przekazane w czasie uruchomienia.
Zaproponuj funkcję, która nie zwróci błędu dla argumentów
3 typów i wyjaśnij jej działanie.

```python
def function_for_3_arg_types(x):
    '''
    Funkcja przyjmuje argument, który może być liczbą całkowitą, ciągiem znaków lub listą. 
    Dla liczby całkowitej zwraca jej dwukrotność.
    Dla ciągu znaków zwraca go napisanego dwa razy pod rząd.
    Dla listy wypisuje wszystkie jej elementy po kolei.
    '''
    if type(x) == int:
        return x * 2
    if type(x) == str:
        return x + x
    if type(x) == list:
        for i in x:
            print(i)
```

```python
x1 = 243
x2 = 'pomidor'
x3 = ['Adam', 34, ['Wojtek']]
```

```python
function_for_3_arg_types(x1)
```

```python
function_for_3_arg_types(x2)
```

```python
function_for_3_arg_types(x3)
```

```python
def function_for_3_arg_types_v2(x):
    '''
    Funkcja działa dla pięciu typów argumentów - int, float, str, list, set.
    Dla int i float zwraca dwukrotność liczby, dla str zwraca str który jest podwojonym podanym str, dla list i set zwraca list albo set który też jest dwukrotnością podanego.
    '''
    return x*2
```

```python
function_for_3_arg_types_v2(x2)
```

# H.I.37

```python
students = [
    {"name": "Alice", "department": "A", "salary": 85},
    {"name": "Bob", "department": "A", "salary": 42},
    {"name": "Charlie", "department": "B", "salary": 85},
    {"name": "David", "department": "C", "salary": 112},
    {"name": "Eve", "department": "B", "salary": 50},
    {"name": "Fox", "department": "C", "salary": 34},
]
```

```python
def create_name_salary_dict(x, name_salary_dict):
    name = x['name']
    salary = x['salary']
    name_salary_dict[name] = salary

def get_highest_salary_and_person(name_salary_dict):
    h_name = ''
    h_salary = 0
    for name, salary in name_salary_dict.items():
        if salary > h_salary:
            h_salary = salary
            h_name = name
    return h_name, h_salary
```

```python
d = {}
for i in students:
    create_name_salary_dict(i, d)
y, z = get_highest_salary_and_person(d)
print(y, z)
```

```python
def counting_numbers(n):
    str_n = str(n)
    # counts_dict = {i:0 for i in range(10)}
    previous = '0'
    num_count = 0
    new_str_n = ''
    for i in str_n:
        if previous != i:
            if previous != '0':
                sufix = str(num_count) + previous
                new_str_n += sufix

            previous = i
            num_count = 0
            num_count += 1

        elif previous == i:
            num_count += 1
    sufix = str(num_count) + previous
    new_str_n += sufix
    return int(new_str_n)
```

```python
def calculate_n_number(x, n):
    current = x
    for _ in range(n):
        current = counting_numbers(current)
    return current
```

```python
x = 1
print(calculate_n_number(x, 1))
print(calculate_n_number(x, 2))
print(calculate_n_number(x, 3))
print(calculate_n_number(x, 23))
```
