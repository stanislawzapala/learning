Lab G.31

```python
import time
import random
def finish_time(start):
    end = time.perf_counter()
    print(f'Czas operacji: {end-start}')
```

```python
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 42},
    {"name": "Charlie", "score": 67},
    {"name": "David", "score": 30},
    {"name": "Eve", "score": 90}
]
```

```python
def find_students_with_score_higher_than(min_score, students):
    found_students = []
    for student in students:
        if student['score'] > min_score:
            found_students.append(student['name'])
    return found_students
    
```

```python
start = time.perf_counter()
find_students_with_score_higher_than(60, students)
finish_time(start)
```

```python
def find_v2(min_score, students):
    return [student['name'] for student in students if student['score'] > min_score]
```

```python
start = time.perf_counter()
find_v2(60, students)
finish_time(start)
```

Lab G.32

```python
def funk1(tab1, tab2):
    if len(tab1) != len(tab2):
        raise ValueError('Tablice muszą być tej samej długości!')
    else:
        sum = 0
        for i in range(len(tab1)):
            sum = sum + tab1[i]*tab2[i]
        return sum
```

```python
tab1 = [i for i in range(10000000)]
tab2 = [1,2,3,4,6]

start = time.perf_counter()
print(funk1(tab1, tab1))
finish_time(start)
```

Lab G.34

```python
def funk_rekur(n):
    if n == 0:
        return 1
    else:
        return n * funk_rekur(n-1)
```

```python
print(funk_rekur(5))
```

Problem - sum of digits in a number

```python
def sum_of_digit( n ):
    if n == 0:
        return 0
    return (n % 10 + sum_of_digit(int(n / 10)))

num = 12345
result = sum_of_digit(num)
print(result)
```

Lab G.35

```python
def find_num(t, c):
    best_num = 0
    dist = abs(best_num - c)
    for i in t:
        new_dist = abs(i - c)
        if  new_dist < dist:
            best_num = i
            dist = new_dist
    return best_num
```

```python
t = [random.randint(1, 1000) for _ in range(1000000)]
c = 140
```

```python
start = time.perf_counter()
print(find_num(t,c))
finish_time(start)
```

```python
# better practices
def find_num2(t, c):
    return min(t, key = lambda x: abs(x-c))
```

```python
start = time.perf_counter()
print(find_num2(t,c))
finish_time(start)
```
