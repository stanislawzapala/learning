## H.I.56

```python
def sprawdz_czy_permutacja(x):
    set_x = set(x)
    if len(set_x) == len(x):
        return 'TAK'
    else:
        return "NIE"
```

```python
lista_1 = ['python', 'java', 'c++', 'rust', 'go']
lista_2 = ['dane', 'algorytm', 'zmienna', 'funkcja', 'dane']
print(sprawdz_czy_permutacja(lista_1))
print(sprawdz_czy_permutacja(lista_2))
```

## H.I.57

```python
def count_sort(x):
    lista_wystapien = []
    for i in range(max(x)+1):
        lista_wystapien.append(0)
    for i in x:
        lista_wystapien[i] += 1
    
    posortowana_lista = []
    
    for i, j in enumerate(lista_wystapien):
        if j != 0:
            for _ in range(j):
                posortowana_lista.append(i)
    
    return posortowana_lista
```
