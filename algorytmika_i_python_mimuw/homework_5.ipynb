### Co robi ten algorytm



```python
def proc(a):
    n = len(a)
    for i in range(n-1):
        zm = False
        for j in range(n-1-i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                zm = True
        if not zm:
            return
```
