## Selection Sort

```python
def selection_sort(set_x):
    x = list(set_x)
    
    if len(x) <= 1:
        return x
    
    min_idx = 0
    for i in range(1, len(x)):
        if x[i] > x[min_idx]:
            min_idx = i
            
    min_value = x.pop(min_idx)
    
    return [min_value] + selection_sort(x)
```

```python
tablica = {15,5,10,9,6,1,2,8,3,4,11}
selection_sort(tablica)
```
